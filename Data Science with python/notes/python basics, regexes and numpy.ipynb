# Programming Language: Objects and map()

```python
class Person: 
    department = "School of Information"
    def set_name(self, new_name):
        self.name = new_name
    def set_location(self, new_location):
        self.location = new_location
```

```python
person = Person()
person.set_name("Chistopher Brooks ")
person.set_location("maaimahiu , Naivasha, Kenya")
print("{}lives in {} and works in the department of {}".format(person.name, person.location, person.department))
```

```python
#finding the minimum value looking through two stores
store1 = [10,11,12.34,2.34]
store2 = [5, 11.10, 12.34, 2.04]
cheapest = map(min, store1, store2)
cheapest
```

```python
people = ["Dr. Christopher Brooks", "Dr. Kevyn Mahugu", "Dr. Ongaro", "Dr. Wambui"]
def split_title_and_name(person):
    title = person.split()[0]
    lastname = person.split()[-1]
    return "{} {}".format(title, lastname)
list(map(split_title_and_name, people))
```

# Patterns and Character Classes

```python
grades = "ACAAAABCBCBAA"
#if we want to find how many B's were in the grade list, we can do the following
re.findall("B",grades)
```

```python
# if we wanted to count the numbers of A's and B's in the list, we con't use "AB" since this is used to match
# all A's followed immediately by a B. Instead we put the characters A and B inside square brackets
re.findall("[AB]", grades)
```

```python
# A simple regex to parse out all instances where this student receives an A followed by a B or a C
re.findall("[A][BC]", grades)
# I think [AB] and [A-B] mean the same thing
```

```python
# another way to do the same thing is to use the pipe operator.
re.findall("AB|AC", grades)
```

```python
# We can use the caret and set operator to negate our results. For instance if we want to parse out only the grades which were not A
re.findall("[^A]", grades)
```

```python
# lets compare the results above with the following
re.findall("^[^A]", grades)
# Observe that we are getting an empty string as a result because  the regex says that we want to match any value at the beginning of the string which is not an A. our string though starts with an A. so there is no match.
```

# QUANTIFIERS

```python
 #quantifiers are the number of times you want a pattern to be matched in order to match.
 # How many times has the student been on a back to back A's streak?
 re.findall("A{2,10}", grades) # 2 is the minimum amount and 10 is max
```

```python
# this is kinda different from the one above. the regx is going to produce instances when the student got 2A\s back to back exactly
re.findall("A{1,1}A{1,1}", grades)
```

```python
re.findall("A{2,2}", grades)
#another way of doing it coz we want to return exactly 2 A's
```

```python
# or we could also do
re.findall("AA", grades)
```

```python
# or we could use one number in the indexing stage and the regex will assume that the given number is the minimum and the max. 
re.findall("A{2}", grades)
```

```python
re.findall("A{1,10}B{1,10}C{1,10}", grades)
```
